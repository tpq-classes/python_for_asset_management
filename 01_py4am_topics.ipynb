## Expected Return

```python
def ER(x0, x1):
    return np.dot(P, x1) - x0
```

```python
ER(S0, S1)
```

```python
def mu(x0, x1):
    return (np.dot(P, x1) - x0) / x0
```

```python
mu(S0, S1)
```

## Volatility

```python
def r(x0, x1):
    return (x1 - x0) / x0
```

```python
r(S0, S1)
```

```python
mu = np.dot(P, r(S0, S1))
```

```python
mu
```

```python
def sigma2(P, r, mu):
    return np.dot(P, (r - mu) ** 2)
```

```python
sigma2(P, r(S0, S1), mu)
```

```python
def sigma_(P, r, mu):
    return np.sqrt(sigma2(P, r(S0, S1), mu))
```

```python
sigma_(P, r, mu)
```

```python
def sigma(P, r, mu):
    return np.sqrt(np.dot(P, (r - mu) ** 2))
```

```python
sigma(P, r(S0, S1), mu)
```

## Optimal Portfolios

```python
B = (10, (11, 11))
```

```python
S = (10, (20, 5))
```

```python
M0 = np.array((B[0], S[0]))
```

```python
M = np.array((B[1], S[1])).T
```

```python
p = 0.5
```

```python
P = np.array((p, 1-p))
```

## Expected Utility

```python
def U(phi):
    c1 = np.dot(M, phi)
    return -np.dot(P, np.log(c1))
```

```python
-U((1, 0))  # all bond
```

```python
-U((0, 1))  # all stock
```

```python
-U((0.5, 0.5))  # mixed bond & stock
```

```python
from scipy.optimize import minimize
```

```python
w = 10
```

```python
cons = ({'type': 'eq',
         'fun': lambda phi: np.dot(M0, phi) - w})  # budget constraint
```

```python
opt = minimize(U, (1, 1), constraints=cons)
```

```python
opt
```

```python
opt['x']  # optimal portfolio
```

```python
-opt['fun']  # maximum expected utility
```

```python
-U(opt['x'])
```

```python
np.dot(M, opt['x'])  # future payoff given optimal portfolio
```

## Time-Additive Expected Utility

```python
M0 = np.array((1, B[0], S[0]))
```

```python
kappa = 10 / 11
```

```python
def U(phi):
    c0 = phi[0]
    c1 = np.dot(M, phi[1:])
    return -(np.log(c0) + kappa * np.dot(P, np.log(c1)))
```

```python
opt = minimize(U, (1, 1, 1), constraints=cons)
```

```python
opt
```

```python
-opt['fun']
```

```python
opt['x'][0]  # cash/money today
```

```python
np.dot(M, opt['x'][1:])  # payoff tomorrow
```

## Diversification

### Risk-Return 

```python
def port_return(phi):
    return np.dot(R.mean(axis=0), phi)  # portfolio return
```

```python
def port_risk(phi):
    return np.dot(phi, np.dot(np.cov(R.T, ddof=0), phi)) ** 0.5  # portfolio volatility
```

```python
M0 = np.array((10, 10))
```

```python
p = np.random.random((500, 2))  # random portfolio compositions ...
p = (p.T / p.sum(axis=1)).T  # ... normalized to 100% (1.0)
p[:7]
```

### Zero Covariance

```python
M = np.array((
    (11, 11),
    (20, 5)
)).T
```

```python
R = M / M0 - 1
R
```

```python
cov = np.cov(R.T, ddof=0)
cov
```

```python
rr = np.array([(port_risk(phi), port_return(phi)) for phi in p])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(rr[:, 0], rr[:, 1], 'ro', label='risk-return')
plt.legend(loc=0)
plt.xlabel('$\sigma_p$')
plt.ylabel('$\mu_p$');
```

### Low Positive Covariance

```python
M = np.array((
    (20, 5),
    (11, 9)
)).T
M
```

```python
R = M / M0 - 1
R
```

```python
cov = np.cov(R.T, ddof=0)
cov
```

```python
corr = cov[0, 1] / (R[:, 0].std() * R[:, 1].std())
corr
```

```python
rr = np.array([(port_risk(phi), port_return(phi)) for phi in p])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(rr[:, 0], rr[:, 1], 'ro', label='risk-return')
plt.legend(loc=0)
plt.xlabel('$\sigma_p$')
plt.ylabel('$\mu_p$');
```
