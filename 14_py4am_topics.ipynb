## Example (1)

### Sample Data

```python
noa = 3
```

```python
mu = np.ones(noa)
mu
```

```python
cov = np.array((
    [9.1, 3.0, 6.0],
    [3.0, 1.1, 2.0],
    [6.0, 2.0, 4.1]
))
cov
```

### Views

#### Basics

```python
v = 1  # number of views
```

```python
P = pd.DataFrame(np.zeros((v, noa)), columns=list('abc'))
P
```

```python
q = np.zeros(v)
q
```

```python
omega = np.zeros((v, v))
omega
```

#### Specific View

```python
P.loc[0, 'a'] = 1
```

```python
P.loc[0, 'b'] = -1
```

```python
q[0] = 2.0
```

```python
omega[0, 0] = 0.00001
```

```python
P
```

### Calculations

```python
m1 = np.dot(P.T, np.dot(1 / omega, P)) + np.linalg.inv(cov)
```

```python
m1
```

```python
m2 = np.dot(P.T, np.dot(1 / omega, q)) + np.dot(np.linalg.inv(cov), mu)
```

```python
m2
```

```python
mu_ = np.dot(np.linalg.inv(m1), m2)
```

```python
mu_  # blended returns
```

```python
r = pd.DataFrame({'hist': mu, 'bl92': mu_}, index=list('abc'))
```

```python
r
```

```python
cov_ = np.linalg.inv(m1)  # blended covariance matrix
```

```python
cov_
```

```python
cov_ + cov
```
